```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
#print()
height = [0.81, 0.71, 0.60, 0.92, 0.98, 0.95, 0.92,]
bars = ('KNN', 'SVM', 'GNB', 'DT', 'XGB', 'CNN', 'MLP')
y_pos = np.arange(len(bars))
```

```python
plt.figure(figsize=(10,8))
plt.bar(y_pos, height, color=[ 'red','orange','yellow', 'green','cyan', 'blue', 'purple'])
plt.xticks(y_pos, bars)
plt.ylabel("Accuracy", fontsize=12)
for index, value in enumerate(height):
    plt.text(index, value, str(value),fontsize = 12)
plt.show()
print("KNN = K-Nearest Neighbor")
print("SVM = Support Vector Machine")
print("GNB = Gaussian Naive Bayes")
print("DT = Desicion Tree")
print("XGB = XgBoost")
print("CNN = Convolutional Neural Network ")
print("MLP = Multi Layer Perceptrons")
```
